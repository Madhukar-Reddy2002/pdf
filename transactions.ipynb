```python
import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("./transactions.csv")
data.head()
```

```python
# Plot the transactions by date
plt.figure(figsize=(12, 6))
sns.barplot(x='date', y='amount', data=data)
plt.title('Transactions by Date')
plt.xlabel('Date')
plt.ylabel('Amount')
plt.show()
```

```python
# Plot the transactions by payee
plt.figure(figsize=(12, 6))
sns.barplot(x='payee', y='amount', data=data)
plt.title('Transactions by Payee')
plt.xlabel('Payee')
plt.ylabel('Amount')
plt.xticks(rotation=90)
plt.show()
```

```python
# Plot the transactions by type
plt.figure(figsize=(8, 6))
sns.countplot(x='type', data=data)
plt.title('Transactions by Type')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```
