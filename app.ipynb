```python
import PyPDF2
import re

def extract_text_without_emojis(pdf_file_path, page_number):
    """Extracts text from a specific page of a PDF file and removes emojis."""
    with open(pdf_file_path, 'rb') as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)
        page = pdf_reader.pages[page_number]
        page_text = page.extract_text()
        clean_text = re.sub(r'[^\x00-\x7F]+', '', page_text)  # Remove non-ASCII, commonly emojis
    return clean_text

def process_transactions(text):
    """Extract transaction details from text using a regex pattern."""
    # Regex pattern to capture date, time, type, amount, and payee details
    pattern = r"(\bJan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)\s(\d{1,2}),\s(\d{4})\s(\d{2}:\d{2}\s(?:AM|PM))\s(DEBIT|CREDIT)\s(\d+,\d+|\d+)\s(Paid to|Received from)(.*)"
    matches = re.findall(pattern, text)
    data_list = []

    for transaction in matches:
        data = {
            "Month": transaction[0],
            "Date": transaction[1],
            "Year": transaction[2],
            "Time": transaction[3],
            "Type": transaction[4],
            "Amount": transaction[5],
            "Transaction": transaction[6],
            "Payee": transaction[7].strip()
        }
        data_list.append(data)

    return data_list

def main():
    # Path to the PDF file
    pdf_file_path = './PhonePe_Statement_Dec2023_Mar2024.pdf'
    # Extract text from the sixth page (index 5) without emojis
    page_text = extract_text_without_emojis(pdf_file_path, 5)

    # Pre-process text if necessary (e.g., removing blank lines or combining lines)
    text_lines = page_text.split('\n')
    text_filtered = " ".join([line.strip() for line in text_lines if line.strip()])

    # Process extracted text to find transaction details
    transactions = process_transactions(text_filtered)

    # Print extracted data
    for transaction in transactions:
        print(transaction)

if __name__ == "__main__":
    main()
```
